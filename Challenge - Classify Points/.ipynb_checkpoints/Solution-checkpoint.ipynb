```python
trainX = pd.read_csv('Logistic_X_Train.csv')
trainY = pd.read_csv('Logistic_Y_Train.csv')
```

```python
X = trainX.values
Y = trainY.values
```

```python
X
```

```python
plt.style.use('seaborn')
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
class Feedforward(torch.nn.Module):
        def __init__(self, input_size, hidden_size, hidden_size_1):
            super(Feedforward, self).__init__()
            self.input_size = input_size
            self.hidden_size  = hidden_size
            self.hidden_size_1 = hidden_size_1
            self.fc1 = torch.nn.Linear(self.input_size, self.hidden_size)
            self.fc2 = torch.nn.Linear(self.hidden_size, self.hidden_size_1)
            self.fc3 = torch.nn.Linear(self.hidden_size_1, 1)
            self.relu = torch.nn.ReLU()
            self.sigmoid = torch.nn.Sigmoid()
        def forward(self, x):
            hidden1 = self.fc1(x)
            relu = self.relu(hidden1)
            hidden2 = self.fc2(relu)
            relu1 = self.relu(hidden2)
            output = self.fc3(relu1)
            output = self.sigmoid(output)
            return output
```

```python
X = torch.FloatTensor(X)
Y = torch.FloatTensor(Y)
```

```python
Y = Y.reshape((Y.shape[0]))
```

```python
print(X.shape, Y.shape)
```

```python
testX = pd.read_csv('Logistic_X_Test.csv')
```

```python
X_ = testX.values
```

```python
X_ = torch.FloatTensor(X_)
```

```python
X_.shape
```

```python
model = Feedforward(2, 10, 10)
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
model.train()
epoch = 20000
losses = []
for epoch in range(epoch):
    optimizer.zero_grad()
    # Forward pass
    y_pred = model(X)
    # Compute Loss
    print(y_pred.shape)
    loss = criterion(y_pred.squeeze(), Y)
    print('Epoch {}: train loss: {}'.format(epoch, loss.item()))
    losses.append(loss.item())
    # Backward pass
    loss.backward()
    optimizer.step()
```

```python
plt.plot(losses)
plt.show()
```

```python
model.eval()
y_pred = model(X_)
```

```python
y_pred
```

```python
f = []
for i in range(y_pred.shape[0]):
    if y_pred[i] >= 0.5:
        f.append(1);
    else: f.append(0)
f = np.array(f)
```

```python
df = pd.DataFrame(f)
```
