```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import cv2
%matplotlib inline
```

# Importing images

```python
img1 = cv2.imread(r"D:\ML\Quickstart Assignment\Quickstart Assignment\top_left.jpg")
img2 = cv2.imread(r"D:\ML\Quickstart Assignment\Quickstart Assignment\top_right.jpg")
img3 = cv2.imread(r"D:\ML\Quickstart Assignment\Quickstart Assignment\bottom_left.jpg")
img4 = cv2.imread(r"D:\ML\Quickstart Assignment\Quickstart Assignment\bottom_right.jpg")
img5 = cv2.imread(r"D:\ML\Quickstart Assignment\Quickstart Assignment\center.jpeg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)
plt.imshow(img5)
```

# Resizing Images

```python
print(type(img1))
img1.shape
```

```python
img1 = cv2.resize(img1, (200, 200))
img2 = cv2.resize(img2, (200, 200))
img3 = cv2.resize(img3, (200, 200))
img4 = cv2.resize(img4, (200, 200))
img5 = cv2.resize(img5, (100, 100))
plt.imshow(img5)
```

# Adding Borders

```python
img1 = cv2.copyMakeBorder(img1, 10, 5, 10, 5, cv2.BORDER_CONSTANT)
img2 = cv2.copyMakeBorder(img2, 10, 5, 5, 10, cv2.BORDER_CONSTANT)
img3 = cv2.copyMakeBorder(img3, 5, 10, 10, 5, cv2.BORDER_CONSTANT)
img4 = cv2.copyMakeBorder(img4, 5, 10, 5, 10, cv2.BORDER_CONSTANT)
img5 = cv2.copyMakeBorder(img5, 10, 10, 10, 10, cv2.BORDER_CONSTANT)
plt.imshow(img4)
```

# Create Collage

```python
upper = np.hstack([img1, img2])
lower = np.hstack([img3, img4])
full = np.vstack([upper, lower])
# plt.imshow(upper)
plt.imshow(full)
```

```python
full[155:275,155:275]=img5[:,:,:]
plt.imshow(full)
```

# Converting image to csv

```python
arr = np.array(full)
# print(arr)
plt.imshow(full)
df=pd.DataFrame(arr.reshape(-1,3),columns=['r','g','b'])
df.to_csv('instagram_collage.csv',index=False)
```
