```python
def odd_one_out(words,word_dict):
    
    all_word_vectors=[word_dict[w.lower()] for w in words]
    avg_vector = np.mean(all_word_vectors,axis=0)
    
    #print(avg_vector.shape)
    
    odd_one_out=None
    min_sim=100.0 # very high value
    
    for w in words:
        sim=cosine_similarity([word_dict[w.lower()]],[avg_vector])
        if sim<min_sim:
            min_sim = sim
            odd_one_out = w
        #print("similarity btw %s and avg vector is %.2f"%(w,sim))
        
    return odd_one_out
```
